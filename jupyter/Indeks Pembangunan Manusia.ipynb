```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_jabar = df_ipm.groupby('nama_kabupaten_kota')
df_jabar
df_jabar_ipm = df_jabar.agg({'indeks_pembangunan_manusia':'mean'})
df_jabar_ipm
```

```python
df_jabar_ipm['ranking'] = df_jabar_ipm['indeks_pembangunan_manusia'].rank(ascending=False)
df_jabar_ipm['ranking_dari_bawah'] = df_jabar_ipm['indeks_pembangunan_manusia'].rank(ascending=True)
df_jabar_ipm = df_jabar_ipm.reset_index()
df_jabar_ipm
```

```python
df_jabar_ipm['ranking'] = df_jabar_ipm['indeks_pembangunan_manusia'].rank(ascending=False)
```

```python
df_ipm_terbawah = df_jabar_ipm[df_jabar_ipm['ranking_dari_bawah']<=5].sort_values(by='ranking',ascending=False)
df_ipm_terbawah
```

```python
df_ipm_terbaik = df_jabar_ipm[df_jabar_ipm['ranking']<=5].sort_values(by='ranking_dari_bawah')
df_ipm_terbaik
```

```python
df_ipm_terbaik.to_csv('ipm_jabar.csv')
```

```python
df_ipm_terbawah.to_csv('ipm_jabar_terbawah.csv')
```

```python
def bodebek(x):
    if x in ['KOTA DEPOK', 'KOTA BOGOR', 'KOTA BEKASI', 'KABUPATEN BOGOR', 'KABUPATEN BEKASI']:
        return 'bodebek'
    else :
        return 'others'
```

```python
df_bodebek_ipm = df_ipm.copy()
df_bodebek_ipm['bodebek'] = df_bodebek_ipm['nama_kabupaten_kota'].apply(bodebek)
df_bodebek_ipm
df_bodebek_ipm = df_bodebek_ipm[df_bodebek_ipm['bodebek']=='bodebek']
df_bodebek_ipm
```

```python
kota_depok = df_ipm[df_ipm['nama_kabupaten_kota']=='KOTA DEPOK']
kota_bogor = df_ipm[df_ipm['nama_kabupaten_kota']=='KOTA BOGOR']
kota_bekasi = df_ipm[df_ipm['nama_kabupaten_kota']=='KOTA BEKASI']
kab_bogor = df_ipm[df_ipm['nama_kabupaten_kota']=='KABUPATEN BOGOR']
kab_bekasi = df_ipm[df_ipm['nama_kabupaten_kota']=='KABUPATEN BEKASI']
```

```python
x = df_ipm['tahun'].unique()
y1 = kota_depok['indeks_pembangunan_manusia']
y2 = kota_bogor['indeks_pembangunan_manusia']
y3 = kota_bekasi['indeks_pembangunan_manusia']
y4 = kab_bogor['indeks_pembangunan_manusia']
y5 = kab_bekasi['indeks_pembangunan_manusia']

plt.plot(x, y1, label='Kota Depok', color='blue')
plt.plot(x, y2, label='Kota Bogor', color='green')
plt.plot(x, y3, label='Kota Bekasi', color='black')
plt.plot(x, y4, label='Kabupaten Bogor', color='red')
plt.plot(x, y5, label='Kabupaten Bekasi', color='brown')

plt.legend()
plt.show
```
